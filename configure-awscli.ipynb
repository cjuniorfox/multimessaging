```python
import subprocess

aws=['aws',
     '--no-sign-request', 
     'sqs', 
     'send-message', 
     '--region', 'sa-east-1', 
     '--endpoint-url', 'http://localhost:4566',
     '--queue-url', 'http://sqs.us-east-1.localhost.localstack.cloud:4566/000000000000/test-queue',
     '--output', 'table']

for i in range(0,100):
    subprocess.Popen(aws + ['--message-body', f"Hello from AWS CLI {i}"])
```
